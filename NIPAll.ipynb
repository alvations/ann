```python
import pickle
from itertools import chain
from collections import namedtuple

import pandas as pd
```

```python
Event = namedtuple('Event', 'id, card, abstract, authors, title, datetime, etype')
```

```python
with open('nips.pk', 'rb') as fin:
    events = pickle.load(fin)
```

```python
events_frame = pd.DataFrame(list(events.values()), columns=Event._fields)
```

```python
events_frame = events_frame.drop_duplicates(['title', 'authors'], take_last=True)
```

```python
tutorials = events_frame[events_frame.etype == "Tutorial"]
```
